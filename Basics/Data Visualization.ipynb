# Data Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.arange(0,10)
```

```python
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
```

```python
plt.plot(x,y,'r-')
```

```python
plt.plot(x,y,'r--')
plt.xlim(2,4)
plt.ylim(10,20)
```

```python
mat = np.arange(0,100).reshape(10,10)
```
